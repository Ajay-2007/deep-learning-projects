```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
plt.style.use('Solarize_Light2')
```

```python
np.random.seed(29)
x = np.random.randn(200)
```

```python
data = pd.DataFrame([x]).T
data.columns = ['x']
```

```python
sns.distplot(data['x'], bins=30);
```

```python
stats.probplot(data['x'], dist='norm', plot=plt)
plt.show()
```

# Real World Example

```python
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()

boston = pd.DataFrame(boston_dataset.data,
                      columns=boston_dataset.feature_names)

boston.head()
```

```python
print(boston_dataset.DESCR)
```

```python
sns.distplot(boston['RM'], bins=30);
```

Normality can be also accessed by Q-Q plots. In a Q-Q plot we plot the quantiles of the variable in the y-axis and the expected quantiles of the normal distribution in the x-axis.
If the variable follows a normal distribution, the dots in the Q-Q plot should fall in a 45 degree diagonal line as indicated below.

```python
stats.probplot(boston['RM'], dist='norm', plot=plt);
plt.show();
```

Most of the observations of RM fall on the 45 degree line, which suggests that the distribution is approximately Gaussian, with some deviation towards the large and smaller values of the variable.

```python
stats.probplot(boston['CRIM'], dist='norm', plot=plt);
plt.show();
```

CRIM feature does not follow a Gaussian distribution as most of its observations deviate from the 45 degree line in the Q-Q plot.
